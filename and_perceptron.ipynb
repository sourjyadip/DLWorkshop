```
import numpy as np
import matplotlib.pyplot as plt
```

```
#Perceptron to simulate AND gate
x1=np.float32(np.array([0.,0.,1.,1.]))
x2=np.float32(np.array([0.,1.,0.,1.]))
yd=np.float32(np.array([0.,0.,0.,1.]))
w1=np.float32(0.5)
w2=np.float32(-0.8)
learn=np.float32(0.1)
no_examples=4
no_epochs=20
err=[]

for j in range(no_epochs):
  for i in range(no_examples):
    summ=(x1[i]*w1)+(x2[i]*w2)
    if summ>=np.float32(0.2):
      y_act=np.float32(1.)
    else:
      y_act=np.float32(0.)
    err.append(yd[i]-y_act)
    del_w1=learn*x1[i]*(yd[i]-y_act)
    del_w2=learn*x2[i]*(yd[i]-y_act)
    w1=w1+del_w1
    w2=w2+del_w2

epochs=np.linspace(1,no_epochs,len(err))
plt.plot(epochs,err)
```
